```python
class BgNode:
    def __init__(self,d,l,r,n):
        self.data = d
        self.left = l
        self.right = r
        self.mid = n
        self.mult = 0
          
    # prints the node and all its children in a string
    def __str__(self):  
        st = "("+str(self.data)+", "+str(self.mult)+") -> ["
        if self.left != None:
            st += str(self.left)
        else: st += "None"
        if self.mid != None:
            st += ", "+str(self.mid)
        else: st += ", None"
        if self.right != None:
            st += ", "+str(self.right)
        else: st += ", None"
        return st + "]"


class BigTree:        
    def __init__(self):
        self.root = None
        self.size = 0

    def __str__(self):
        return str(self.root)

    def add(self, st):
        if st == "":
            return None
        if self.root == None:
            self.root = BgNode(st[0],None,None,None)
        ptr = self.root
        for i in range(len(st)):
            d = st[i]
            while True:
                if d == ptr.data:
                    break
                elif d < ptr.data:
                    if ptr.left == None:
                        ptr.left = BgNode(d,None,None,None)
                    ptr = ptr.left
                else:
                    if ptr.right == None:
                        ptr.right = BgNode(d,None,None,None)
                    ptr = ptr.right
            if i < len(st)-1 and ptr.mid == None:
                ptr.mid = BgNode(st[i+1],None,None,None)
            if i < len(st)-1:
                ptr = ptr.mid
        ptr.mult += 1
        self.size += 1

    def addAll(self, A):
        for x in A: self.add(x)
        
    def printAll(self):
        def printFrom(ptr,s):
            if ptr == None: return
            s0 = s + ptr.data
            if ptr.left != None: printFrom(ptr.left,s)
            for i in range(ptr.mult): print(s0, end=" ")
            if ptr.mid != None: printFrom(ptr.mid,s+ptr.data)
            if ptr.right != None: printFrom(ptr.right,s)

        printFrom(self.root,""); print()      
        


    def count(self, st):
        # TODO: should count how many times st appears in the tree
        if st == None or st == "": 
            ### if the count function is used with an empty string or a NULL value, 
            ### it should not change Tree and return None
            return None
        
        ### internal function for Recursion
        def countRecur(BGnode, strFind, fooWord):
            _next = strFind[len(fooWord)]
            ### _next - the next character from the string to search in the TREE

            if BGnode.data == _next: ### when character on node matches _next, 
                fooWord += BGnode.data ### add it to the fooWord - a temporary word to keep track of the word in the TREE
                
                if fooWord == strFind: ### string is found when fooWord = strFind
                    return BGnode.mult ### return multiplicity - no. of occurences
                else:
                    ### move onto the next character, the mid, if it exists
                    if (BGnode.mid != None):
                        BGnode = BGnode.mid
                        return countRecur(BGnode, strFind, fooWord)
                    else:
                        ### string isn't there in the TREE
                        return 0

            else:
                ### if it doesn't match, search on left and then right
                if _next < BGnode.data:
                    if (BGnode.left != None):
                        BGnode = BGnode.left
                        return countRecur(BGnode, strFind, fooWord)
                    else:
                        ### isn't there on left
                        return 0
                else:
                    if (BGnode.right != None):
                        BGnode = BGnode.right
                        return countRecur(BGnode, strFind, fooWord)
                    else:
                        ### isn't there on left
                        return 0
        totalNoOfReps = countRecur(self.root,st,"")
        return totalNoOfReps
    
    
    ### for using in the toIncArray function
    def append(A, k):
        B = [0 for i in range(len(A)+1)]
        for i in range(len(A)):
            B[i] = A[i]
        B[len(A)] = k
        return B


    def toIncArray(self):
         # TODO: should return an array with all the elements in increasing order
        
        ### start with left, since we need go ascending and left has the smaller values
        def inRec (BGnode, fooArray, fooWord):
            if BGnode.left: ### traverses the left sub-TREE of the node, if it exists
                fooArray = inRec(BGnode.left, fooArray, fooWord) ### array holds the words found, if any

            if BGnode.mult > 0: ### if data forms a word (mult == 1 or mult == 2), add to array
                for i in range (BGnode.mult):
                    fooArray = BigTree.append(fooArray, fooWord+BGnode.data)

            ### before traversing right, middle should come as it has greater priority than right for ASCENDING, small --> big
            if BGnode.mid: ### traverses node pointed by the mid, if it exists
                fooArray = inRec(BGnode.mid, fooArray, fooWord+BGnode.data) ### array holds the words found if any


            if BGnode.right: ### traverses the right sub-TREE of the node, if it exists
                fooArray = inRec(BGnode.right, fooArray, fooWord)

            return fooArray ### array of all the words in the TREE

        A = inRec(self.root, [], "")
        return A
```
